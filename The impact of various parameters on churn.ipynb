```python
result = pd.pivot_table(data=df, index='Tenure Group', columns='Monthly Charges Group',values='Churn')
result
```

```python
plt.figure(figsize = (10,5))
sns.heatmap(result, annot=True,cmap = sns.cubehelix_palette(as_cmap=True))
plt.xlabel('Monthly Charges Group', fontsize = 12)
plt.ylabel('Tenure Group', fontsize = 12)
plt.tick_params(axis = 'both', which = 'major', labelsize = 10)
plt.text(x=0, y=-0.8, s='How does the monthly charges group and tenure group affect churn?', fontsize=12, weight='bold')
plt.text(x=0, y=-0.5, s='The effect of Churn Rate on Monthly Charges and Tenure where customers', fontsize=12, alpha=0.75)
plt.text(x=0, y=-0.3, s='do not have Partner and Dependents, but choose long-term contracts', fontsize=12, alpha=0.75)
plt.text(x=0, y=-0.1, s='(One year and Two years contract)', fontsize=12, alpha=0.75)
plt.show()
```
